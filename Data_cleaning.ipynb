Agregaciòn de datos por categorìa

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(2021)

n = 500
gender = ['Male', 'Female']
social_status = ['Poor','Middle class', 'Rich']

gender_data = []
social_status_data = []

for i in range(n):
    gender_data.append(np.random.choice(gender))
    social_status_data.append(np.random.choice(social_status))

height = 160 + 30 * np.random.randn(n)
weight = 70 + 25 *np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3000 * np.random.randn(n)

data = pd.DataFrame({'Gender':gender_data,
                     'Social_status': social_status_data,
                     'Height': height,
                     'Weight': weight,
                     'Age': age,
                     'Income': income})
data
```

Agrupaciòn groupBy

```python
grouped_gender = data.groupby('Gender')

for name, gruop in grouped_gender:
    print(name)
    print(gruop)
```

```python
# para obtener el grupo utilizo el metodo get_group

Male_group = grouped_gender.get_group('Male')
Male_group
```
