```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defino la funci칩n que deseo identificar el punto mas bajo
func = lambda th: np.sin((1/2 * th[0]**2) - (1/4 * th[1]**2) + 3 ) * np.cos((2 * th[0]) + 1 - np.e**th[1])

#defino 2 arreglos de -2 a 2 distribuidos en 100 datos para X y Y, ademas una matriz de 0 con el tama침o x*y.
res = 100

_X = np.linspace(-2,2,100)
_Y = np.linspace(-2,2,100)
_Z = np.zeros((res,res))

#recorro la matriz nula _Z y calculo para cada ix,iy el valor correspondiente a la funci칩n definida
for ix, x_value in enumerate(_X):
  for iy, y_value in enumerate(_Y):
    _Z[ix,iy] = func([x_value, y_value])

#establezco un punto aleatorio de partida
theta = np.random.rand(2) * 4 - 2
_T = np.copy(theta)


h = 0.001 # parametro para definir el incremento de la derivada
lr = 0.001 # ratio de aprendizaje

gradiente = np.zeros(2)

plt.plot(theta[0], theta[1], "o", c = 'white')

for _ in range(10000):
  for it, t in enumerate(theta):

    _T = np.copy(theta)

    _T[it] = _T[it] + h

    derivada = (func(_T) - func(theta))/h

    gradiente[it] = derivada

  # calculamos la ubicaci칩n del nuevo punto
  theta = theta - lr * gradiente

  if (_ % 100 == 0):
    plt.plot(theta[0],theta[1], '.', c = 'red')

plt.plot(theta[0], theta[1], "o", c = 'green')



plt.contourf(_X,_Y,_Z, 100 )
plt.title('Descenso del Gradiente')

```
